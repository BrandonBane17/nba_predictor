```python
from nba_api.stats.endpoints import leaguegamefinder
```

```python
gamefinder = leaguegamefinder.LeagueGameFinder(date_from_nullable='10/24/2023', date_to_nullable='4/14/2024', league_id_nullable='00')
games = gamefinder.get_data_frames()[0]
games.head()
```

```python
games.columns
```

```python
games=games[['TEAM_NAME', 'GAME_ID', 'GAME_DATE', 'MATCHUP', 'WL', 'PLUS_MINUS']]
```

```python
games
```

```python
import pandas as pd
```

```python
games.info()
```

```python
games['GAME_DATE']=pd.to_datetime(games['GAME_DATE'])
```

```python
games['GAME_DATE']
```

```python
games=games.sort_values('GAME_DATE')
```

```python
games
```

```python
games['avg_last_30_+-']=games.groupby('TEAM_NAME')['PLUS_MINUS'].transform(lambda x: x.rolling(30,closed='left').mean())
```

```python
games[games['TEAM_NAME']=='Detroit Pistons'].head(82)
```

```python
msk=games['MATCHUP'].str.contains('@')
```

```python
away_game=games[msk]
home_game=games[~msk]
```

```python
home_game.shape
```

```python
away_game.shape
```

```python
home_game
```

```python
away_game
```

```python
games_merged=pd.merge(home_game, away_game,on='GAME_ID',suffixes=('_HOME','_AWAY'))
```

```python
games_merged
```

```python
games_merged['avg_30_+-_diff']=games_merged['avg_last_30_+-_HOME']-games_merged['avg_last_30_+-_AWAY']
```

```python
games_merged
```

```python
games_model=games_merged[['WL_HOME', 'avg_30_+-_diff']].dropna()
```

```python
games_model['WL_HOME']=games_model['WL_HOME'].map({'W':1,'L':0})
```

```python
games_model
```

```python
from sklearn.model_selection import train_test_split
```

```python
%pip install scikit-learn
```

```python
df_train,df_test=train_test_split(games_model,stratify=games_model['WL_HOME'],test_size=0.2,random_state=7)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
target='WL_HOME'
x_train=df_train.drop(columns=target)
y_train=df_train[target]

x_test=df_test.drop(columns=target)
y_test=df_test[target]
```

```python
pip install xgboost
```

```python
import xgboost as xgb
```

```python
clf=xgb.XGBClassifier(random_state=7)
```

```python
clf.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=clf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import loguniform
```

```python
hyp_params={'learning_rate':loguniform(0.0001,1),
            'max_depth':[2,3,4,5,6,7,8,9],
            'subsample':[0.7,0.8,0.9,1.0],
            'n_estimators':[50,100,150,200]}
```

```python
random_hyp=RandomizedSearchCV(estimator=clf,
                              param_distributions=hyp_params,
                              n_iter=20,
                              cv=7,
                              scoring='accuracy',
                              random_state=7)
```

```python
random_hyp.fit(x_train,y_train)
```

```python
random_hyp.best_params_
```

```python
model_hyp=random_hyp.best_estimator_
```

```python
y_pred_hyp=model_hyp.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_hyp)
```

```python
from joblib import dump, load
```

```python
dump(model_hyp,'model_nba.joblib')
```

```python
model_saved=load('model_nba.joblib')
```

```python
accuracy_score(y_test,model_saved.predict(x_test))
```

```python
team_home='Detroit Pistons'
team_away='Golden State Warriors'
```

```python
def predict_games(team_home, team_away):
    from nba_api.stats.endpoints import leaguegamefinder
    gamefinder = leaguegamefinder.LeagueGameFinder(date_from_nullable='10/24/2023', date_to_nullable='4/14/2024', league_id_nullable='00')
    games = gamefinder.get_data_frames()[0]

    games = games[['TEAM_NAME', 'GAME_ID', 'GAME_DATE', 'MATCHUP', 'WL', 'PLUS_MINUS']]

    import pandas as pd

    games['GAME_DATE'] = pd.to_datetime(games['GAME_DATE'])

    msk_home=(games['TEAM_NAME']==team_home)
    games_30_home=games[msk_home].sort_values('GAME_DATE').tail(30)
    home_plus_minus=games_30_home['PLUS_MINUS'].mean()

    msk_away=(games['TEAM_NAME']==team_away)
    games_30_away=games[msk_away].sort_values('GAME_DATE').tail(30)
    away_plus_minus=games_30_away['PLUS_MINUS'].mean()

    games_diff = home_plus_minus - away_plus_minus

    model_saved=load('model_nba.joblib')

    import numpy as np
    predict_home_win=model_saved.predict(np.array([games_diff]))[0]

    predict_winning_probability=model_saved.predict_proba(np.array([games_diff]))[0][1]
    return predict_home_win,predict_winning_probability
```

```python
predict_games('Detroit Pistons','Washington Wizards')
```
